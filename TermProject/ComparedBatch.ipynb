# Plot Graph

```python
# Assuming both datasets have a column named 'Gender' for x-axis values
# and 'Satisfy Score' for y-axis values
x1 = FirstBatch['Satisfy Score']
x2 = SecondBatch['Satisfy Score']

# Plotting the data
plt.figure(figsize=(8, 6))

# Histogram for the 1st Batch with a specific color
plt.hist(x1, bins=20, alpha=0.7, color='blue', label='1st Batch')

# Histogram for the 2nd Batch with a different color
plt.hist(x2, bins=20, alpha=0.7, color='orange', label='2nd Batch')

# Adding labels and title
plt.xlabel('Satisfy Score')
plt.ylabel('Frequency')
plt.title('Histogram of Satisfy Score for Two Batches')
plt.legend()

# Set the x-axis ticks with specific decimal formatting
plt.xticks(list(range(int(min(min(x1), min(x2))), int(max(max(x1), max(x2))) + 1)), rotation=45, ha="right", rotation_mode="anchor", fontsize=8)

# Display the plot
plt.show()
```
